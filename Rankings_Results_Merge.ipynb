```python
import pandas as pd
import numpy as np
```

## Change weight class and restart kernel to select new weight

```python
weight_class = 285
```

```python
rankings = pd.read_csv('2019_Rankings.csv')
```

```python
rankings
```

```python
ind_list = []
for i in range(0,len(rankings['WeightClass'])):
    if rankings['WeightClass'][i] == weight_class:
        ind_list.append(i)
```

```python
rankings = rankings.iloc[ind_list[0]:ind_list[-1]+1]
```

```python
names1 = []
for item in rankings['FirstName']+rankings['LastName']:
    names1.append(item.lower())
rankings['NameLower'] = names1
```

```python
tourn = pd.read_csv('2019_Tournament_Scored/All_{}_scored_df.csv'.format(weight_class))
```

```python
names = []
for item in tourn.iloc[:,1]:
    names.append(item.lower())
tourn['NameLower'] = names
tourn_trunc = tourn.iloc[:,-5:]
```

```python
from fuzzywuzzy import fuzz
```

## Use this cell to set fuzzy_limit to match misspellings properly

```python
fuzzy_limit = 70
```

```python
quals_name_changed = []
print('This output should show only valid name substitutions, adjust fuzzy limit if not.\n')
for item in tourn['NameLower']:
    for item1 in rankings['NameLower']:
        if (fuzz.ratio(item,item1) > fuzzy_limit) & (fuzz.ratio(item,item1) < 100):
            quals_name_changed.append(item)
            print(item,item1,fuzz.ratio(item,item1))
```

```python
#error_starter
```

```python
count = -1
for item in tourn_trunc['NameLower']:
    count = count + 1
    for item1 in rankings['NameLower']:
        if (fuzz.ratio(item,item1) > fuzzy_limit):
            tourn_trunc['NameLower'][count] = item1
```

```python
ranked_ncaa_qualifiers = pd.merge(rankings,tourn_trunc,how='inner',on='NameLower')
```

```python
ranked_ncaa_qualifiers
```

## Which NCAA Qualifiers aren't on rankings list?

```python
quals_not_ranked = []
for item in tourn_trunc['NameLower']:
    if item not in list(ranked_ncaa_qualifiers['NameLower']):
        quals_not_ranked.append(item)
for item in quals_not_ranked:
    if item in quals_name_changed:
        quals_not_ranked.remove(item)
```
